# Visualization of configurations at baited fragments

We implemented the visualization of configurations at baited fragments in triangle interaction maps in the module `TIMviz`. In this representation, interactions are shown as rectangles that are colored according to their configuration. The module is implemented here:

```python
from diachr.TIMViz import TIMViz
```

How to use this module is demonstrated [here](../usage/triangle_interaction_maps.ipynb).

## Input data

### Choose an interaction file

```python
AUTHOR = 'MIF' # JAV, SCH, MIF
PROTOCOL = 'CHC' # HC or CHC
CELL_TYPE_SHORT = 'GM12878' # MK, ERY, NEU, MON, MAC_M0, ..., , MESC, GM12878
OUT_PREFIX = AUTHOR + '_' + CELL_TYPE_SHORT + '_' + PROTOCOL +'_REPC'
INTERACTION_FILE = '../../UICer_interactions/' + PROTOCOL + '/' + OUT_PREFIX + '_evaluated_and_categorized_interactions.tsv.gz' 
```

### Choose a BED file with regions of baited fragments

```python
BAITED_FRAGMENT_FILE = '../../additional_files/javierre_2016/javierre_baited_digests_hg38.bed'
```

### Create a `TIMViz` object

```python
visualizer = TIMViz(i_file=INTERACTION_FILE, b_file=BAITED_FRAGMENT_FILE, verbose=False)
```

## Configuration color assignment

If the visualization function is called with `color_i_cats=False`, then interactions are colored according to their configurations. The color assignment is as follows:

* `01`: pink
* `02`: red
* `03`: lime
* `12`: magenta
* `13`: blue
* `23`: turquoise
* `0X`: lightsteelblue
* `1X`: lightsteelblue
* `2X`: lightsteelblue
* `3X`: lightsteelblue
* Balanced interactions: dimgray

In our analyses, interactions with configurations `0X`, `1X`, `2X`, and `3X` account for only a small proportion of all interactions. To make the visualizations clearer, we assigned the same color to all these configurations. In the case of balanced interactions, the configurations result from small deviations from the uniform distribution of the four counts. Therefore, we have assigned the color gray to all balanced interactions.

## Separate visualizations of interactions going from baited fragments towards 5' and 3'

Interactions that go from baited fragments towards `5'` have enrichment status `NE`  and those that go towards `3'` have enrichment status `EN`. In order to obtain clear representations for a larger section of the genome, we visualized `NE`  and `EN` interactions separately.

We created visualizations for the following region:

```python
chrom = 'chr1'
begin = 23000000
end = 25000000
pdf_file_name_prefix = NOTEBOOK_RESULTS_DIR + '/' + 'TIMviz_ht_conf_' + chrom + ':' + str(begin) + '-' + str(end)
```

In total, we create four visualizations, one for each enrichment state (`NE` and `EN`) and one each for unbalanced (`DIX` and `DI`) and balanced interactions (`UIR` and `UI`).

```python
print('Creating plots for: ' + chrom + ':' + str(begin) + '-' + str(end))
for e_cat in ['NE', 'EN']:
    print("\te_cat: " + e_cat)
    for i_cat_list in [['DIX', 'DI'], ['UIR', 'UI']]:
        print("\t\ti_cat_list: " + str(i_cat_list))
        if i_cat_list == ['DIX', 'DI']:
            plot_title = 'Unbalanced interactions'
        if i_cat_list == ['UIR', 'UI']:
            plot_title = 'Balanced interactions'
        if e_cat == 'NE':
            plot_title = plot_title + ' from the baits towards 5\''
        if e_cat == 'EN':
            plot_title = plot_title + ' from the baits towards 3\''
        TIMviz_plot = visualizer.create_visualization(chrom = chrom,
                                        begin = begin,
                                        end = end,
                                        inter_cat_list = i_cat_list,
                                        enr_cat_list = [e_cat],
                                        color_i_cats=False,
                                        pp_linewidth=0,
                                        transparent_polygons=False,           
                                        plot_title = plot_title,
                                        pdf_file_name = pdf_file_name_prefix + '_' + e_cat + '_' + str(i_cat_list).replace(" ", "")  + '.pdf',
                                        verbose=False)
```

## Joint visualizations of `NE` and `EN` interactions for a smaller section of the genome

We created visualizations for the following region:

```python
chrom = 'chr1'
begin = 23600000
end = 24400000
pdf_file_name_prefix = NOTEBOOK_RESULTS_DIR + '/' + 'TIMviz_ht_conf_' + chrom + ':' + str(begin) + '-' + str(end)
```

In total, we create three visualizations, one for unbalanced interactions only, one for balanced interactions only and one for unbalanced and balanced interactions together.

```python
print('Creating plots for: ' + chrom + ':' + str(begin) + '-' + str(end))
for i_cat_list in [['DIX', 'DI'], ['UIR', 'UI'],['DIX', 'DI', 'UIR', 'UI']]:
    print("\t\ti_cat_list: " + str(i_cat_list))
    if i_cat_list == ['DIX', 'DI']:
        plot_title = 'Unbalanced interactions'
    elif i_cat_list == ['UIR', 'UI']:
        plot_title = 'Balanced interactions'
    elif i_cat_list == ['DIX', 'DI', 'UIR', 'UI']:
        plot_title = 'Unbalanced and balanced interactions'
    plot_title = plot_title + ' from the baits towards 5\' and 3\''
    TIMviz_plot = visualizer.create_visualization(chrom = chrom,
                                    begin = begin,
                                    end = end,
                                    inter_cat_list = i_cat_list,
                                    enr_cat_list = ['NE','EN'],
                                    color_i_cats=False,
                                    ht_tag_list = ['0X','1X','2X','3X','01','02','03','12','13','23'],
                                    pp_linewidth=0,
                                    plot_bait_diagonals=True,
                                    transparent_polygons=False,           
                                    plot_title = plot_title,
                                    show_complete_annotations=False,
                                    pdf_file_name = pdf_file_name_prefix + '_' + str(i_cat_list).replace(" ", "") + '.pdf',
                                    verbose=False)
```

## Sliding a window along the genome

We identified two configuration patterns of baited fragments that are found throughout the genome. To give an impression of the ubiquity of these patterns, we slide a window along the genome.

```python
chrom = 'chr15'
from_pos = 90464732-500000
to_pos = from_pos + 3000000
window_size = 1000000
shift_size = int(window_size/2)
for start_pos in range(from_pos, to_pos, shift_size):
    begin = start_pos
    end = begin + window_size
    print('Creating plot for: ' + chrom + ':' + str(begin) + '-' + str(end))
    TIMviz_plot = visualizer.create_visualization(chrom = chrom,
                                    begin = begin,
                                    end = end,
                                    inter_cat_list = ['DIX', 'DI','UIR', 'UI'],
                                    enr_cat_list = ['NE','EN'],
                                    ht_tag_list = ['02','12','03','13','12','23'],
                                    plot_bait_diagonals=True,
                                    transparent_polygons=False, 
                                    color_i_cats=False,
                                    verbose=False)
```
